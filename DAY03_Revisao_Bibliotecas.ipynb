```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df[['Salary','SalaryUnit']] = df['Salary'].str.split('/',expand=True)
```

```python
df.head()
```

```python
df['Salary'] = df['Salary'].apply(lambda linha: linha.lstrip('₹').lstrip('$').lstrip('£').lstrip('AFN ').replace(',',''))
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Salary'] = df['Salary'].astype(int)
```

```python
df.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df.describe()
```

```python
df['SalaryUnit'].unique()
```

```python
df[df['Salary'] == 26.00]
```

```python
#Resolver salários com unidades diferentes -> HORA - MÊS / ANO - MÊS
```

```python
df = df[(df['SalaryUnit'] != 'mo') & (df['SalaryUnit'] !='hr')]
```

```python
df['SalaryUnit'].unique()
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['Salary'])
```

```python
df_agg = df.groupby('Location')['Salary'].agg('mean').reset_index()
```

```python
sns.barplot(x='Location',y='Salary',data=df_agg)
plt.show()
```

```python
df_IBM = df[df['Company Name']=='IBM']
df_IBM.head()
```

```python
sns.boxplot(x='Job Title',y='Salary',data = df_IBM)
plt.show()
```
